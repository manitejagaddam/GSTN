```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
input = pd.read_csv("X_Train_Data_Input.csv")
output = pd.read_csv("Y_Train_Data_Target.csv")
```

```python
df = pd.merge(input, output)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop("ID", axis = 1)
```

```python
null_percentage = df.isnull().sum()/len(df)*100
```

```python
df.head()
```

```python
df.shape
```

```python
cols = null_percentage[null_percentage < 5].iloc[:].index
```

```python
type(cols)
```

```python
for i in cols:
    print(i, end = " Unique Values : ")
    print(df[i].unique())
    print("Number of unique values : ", end = "")
    print(df[i].nunique())
    print("______________________________________________")
```

```python
df_new = df[cols]
```

```python
df_new.head()
```

```python
df_new.shape
```

```python
df_new.dropna().shape
```

```python
df_new_copy = df_new.copy()
```

```python
df_new = df_new.dropna()
```

```python
df_new_sample = df_new.sample(100000)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
pca = PCA(n_components = 10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_new_sample.drop("target", axis = 1)
y = df_new_sample['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 88)
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_pca = pca.fit_transform(x_train_scaled)
```

```python
x_test_pca = pca.transform(x_test_scaled)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_pca, y_train)
```

```python
y_pred = model.predict(x_test_pca)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred, y_test)
```

```python
accuracy_score(y_test, y_pred)
```
